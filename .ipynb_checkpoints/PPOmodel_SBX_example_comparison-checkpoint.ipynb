```python
import time
import json
import os
import torch
import gymnasium as gym
from stable_baselines3 import PPO as PPO_SB3
from sbx.ppo import PPO as PPO_SBX
import matplotlib.pyplot as plt
import jax.numpy as jnp
import jax

import flax.linen as nn
import optax
from sbx.common.policies import BaseJaxPolicy
from flax.training.train_state import TrainState
from sbx.ppo.policies import Actor, Critic, PPOPolicy
from sbx.common.policies import Flatten
from typing import Optional, Callable, Any
import numpy as np

from gymnasium.envs.registration import register

from gymnasium import spaces
import tensorflow_probability.substrates.jax as tfp

import matplotlib.pyplot as plt
from jax import config
from jax import nn as jnn  # Correct activation functions live here
import torch
```

```python
# Configure JAX for M1/M2 Macs
config.update("jax_platform_name", "cpu")  # Force CPU (Metal support is experimental)

# ===== Configuration =====
ENV_NAME = "Pendulum-v1"  # Lightweight environment
TOTAL_TIMESTEPS = 50_000  # Reduced for quicker testing

# File paths
SB3_MODEL_PATH = "sb3_ppo_model.zip"
SBX_MODEL_PATH = "sbx_ppo_model"
RESULTS_FILE = "ppo_timing_results.json"

# Policy configurations optimized for M2 CPU
SB3_POLICY_KWARGS = {
    "net_arch": dict(pi=[32, 32], vf=[32, 32]),  # Smaller network
    "activation_fn": torch.nn.ReLU  # Use torch's ReLU
}

SBX_POLICY_KWARGS = {
    "net_arch": [32, 32],  # Smaller network for CPU
    "activation_fn": jnp.tanh,  # JAX's tanh
    "ortho_init": False,
    "optimizer_kwargs": {"eps": 1e-5}  # Better stability on CPU
}

def train_sb3():
    """Train and save SB3 model (PyTorch)"""
    print("\n=== Training SB3 (PyTorch) ===")
    try:
        env = gym.make(ENV_NAME)
        
        model = PPO_SB3(
            "MlpPolicy",
            env,
            verbose=1,
            device="cpu",  # Force CPU for fair comparison
            batch_size=128,  # Smaller batches
            n_steps=512,
            n_epochs=5,
            policy_kwargs=SB3_POLICY_KWARGS
        )
        
        start_time = time.time()
        model.learn(total_timesteps=TOTAL_TIMESTEPS)
        training_time = time.time() - start_time
        
        model.save(SB3_MODEL_PATH)
        print(f"SB3 model saved to {SB3_MODEL_PATH}")
        return training_time
        
    except Exception as e:
        print(f"SB3 training failed: {e}")
        return None
    finally:
        env.close()

def train_sbx():
    """Train and save SBX model (JAX CPU)"""
    print("\n=== Training SBX (JAX CPU) ===")
    try:
        env = gym.make(ENV_NAME)
        
        model = PPO_SBX(
            "MlpPolicy",
            env,
            verbose=1,
            batch_size=128,
            n_steps=512,
            n_epochs=5,
            policy_kwargs=SBX_POLICY_KWARGS
        )
        
        start_time = time.time()
        model.learn(total_timesteps=TOTAL_TIMESTEPS)
        training_time = time.time() - start_time
        
        model.save(SBX_MODEL_PATH)
        print(f"SBX model saved to {SBX_MODEL_PATH}")
        return training_time
        
    except Exception as e:
        print(f"SBX training failed: {e}")
        return None
    finally:
        env.close()

def save_results(sb3_time=None, sbx_time=None):
    """Save timing results to JSON"""
    results = {}
    if os.path.exists(RESULTS_FILE):
        with open(RESULTS_FILE, "r") as f:
            results = json.load(f)
    
    if sb3_time is not None:
        results["sb3_time"] = sb3_time
    if sbx_time is not None:
        results["sbx_time"] = sbx_time
    
    with open(RESULTS_FILE, "w") as f:
        json.dump(results, f)

def load_results():
    """Load timing results from JSON"""
    if os.path.exists(RESULTS_FILE):
        with open(RESULTS_FILE, "r") as f:
            return json.load(f)
    return {"sb3_time": None, "sbx_time": None}

def run_comparison():
    """Run and compare both implementations"""
    results = load_results()
    
    # SB3 Training (only if needed)
    if not os.path.exists(SB3_MODEL_PATH):
        print("\nStarting SB3 training...")
        sb3_time = train_sb3()
        if sb3_time is not None:
            save_results(sb3_time=sb3_time)
    else:
        print("\nSB3 model already exists at:", SB3_MODEL_PATH)
        if results.get("sb3_time"):
            print(f"Previous SB3 training time: {results['sb3_time']:.2f}s")
        else:
            # If model exists but no timing data, set dummy time
            results["sb3_time"] = 0
    
    # SBX Training (only if needed)
    if not os.path.exists(SBX_MODEL_PATH):
        print("\nStarting SBX training...")
        sbx_time = train_sbx()
        if sbx_time is not None:
            save_results(sbx_time=sbx_time)
    else:
        print("\nSBX model already exists at:", SBX_MODEL_PATH)
        if results.get("sbx_time"):
            print(f"Previous SBX training time: {results['sbx_time']:.2f}s")
        else:
            # If model exists but no timing data, set dummy time
            results["sbx_time"] = 0
    
    # Load latest results
    results = load_results()
    
    # Display comparison if both results exist
    if results.get("sb3_time") and results.get("sbx_time"):
        print("\n=== Final Results ===")
        print(f"SB3 (PyTorch): {results['sb3_time']:.2f}s")
        print(f"SBX (JAX): {results['sbx_time']:.2f}s")
        if results['sb3_time'] > 0 and results['sbx_time'] > 0:
            print(f"Speedup: {results['sb3_time']/results['sbx_time']:.2f}x")
        
        plt.figure(figsize=(8, 5))
        plt.bar(["SB3 (PyTorch)", "SBX (JAX)"], 
                [results["sb3_time"], results["sbx_time"]],
                color=['blue', 'orange'])
        plt.ylabel("Training Time (seconds)")
        plt.title("PPO Training Time Comparison")
        plt.show()

if __name__ == "__main__":
    run_comparison()
```

```python
import os
import time
import json
import gym
import torch
import jax
import optax
import numpy as np
import matplotlib.pyplot as plt
from jax import numpy as jnp
from flax import linen as jnn
from stable_baselines3 import PPO as PPO_SB3
from sbx.ppo import PPO as PPO_SBX
from jax import config
from gym.wrappers import ClipAction

# ===== Fix numpy bool8 issue =====
np.bool8 = np.bool_  # Workaround for legacy numpy versions

# ===== JAX Configuration =====
config.update("jax_platform_name", "cpu")
config.update("jax_disable_jit", False)
config.update("jax_debug_nans", False)
print("Using JAX CPU backend with optimizations")

# ===== PyTorch Configuration =====
torch.set_num_threads(torch.get_num_threads())
print(f"PyTorch using {torch.get_num_threads()} threads")

# ===== Environment Configuration =====
ENV_NAME = "Pendulum-v1"
TOTAL_TIMESTEPS = 100_000

# File paths
SB3_MODEL_PATH = "sb3_ppo_model.zip"
SBX_MODEL_PATH = "sbx_ppo_model"
RESULTS_FILE = "ppo_timing_results.json"

force = True

# ===== Fixed Hyperparameters =====
SB3_CONFIG = {
    "policy_kwargs": {
        "net_arch": dict(pi=[32,32], vf=[32,32]),
        "activation_fn": torch.nn.ReLU
    },
    "batch_size": 256,
    "n_steps": 1024,
    "n_epochs": 5,
    "gamma": 0.99,
    "gae_lambda": 0.95,
    "clip_range": 0.2,
    "device": "cpu",
    "learning_rate": 3e-4
}

SBX_CONFIG = {
    "learning_rate": 3e-4,
    "gamma": 0.99,
    "gae_lambda": 0.95,
    "clip_range": 0.2,
    "batch_size": 4096,
    "n_steps": 4096,
    "n_epochs": 5,
    "policy_kwargs": {
        "net_arch": [128, 128],
        "activation_fn": jnn.relu,
        "ortho_init": False,
        "optimizer_class": optax.adamw,
        "optimizer_kwargs": {"eps": 1e-5, "weight_decay": 1e-4},
        # Removed dtype parameter
    }
}

# ===== Training Functions ===== 
def train_sb3():
    """Train SB3 model with proper action clipping"""
    print("\n=== Training SB3 (PyTorch CPU) ===")
    try:
        env = gym.make(ENV_NAME)
        env = ClipAction(env)
        env = gym.wrappers.RecordEpisodeStatistics(env)
        
        model = PPO_SB3(
            "MlpPolicy",
            env,
            verbose=0,
            **SB3_CONFIG
        )
        
        print("Running PyTorch warmup...")
        model.learn(total_timesteps=1024)
        
        print("Starting main training...")
        start_time = time.time()
        model.learn(total_timesteps=TOTAL_TIMESTEPS)
        training_time = time.time() - start_time
        
        model.save(SB3_MODEL_PATH)
        print(f"SB3 training completed in {training_time:.2f}s")
        return training_time
    except Exception as e:
        print(f"SB3 training failed: {e}")
        return None
    finally:
        env.close()

def train_sbx():
    """Train SBX model with JIT warmup"""
    print("\n=== Training SBX (JAX CPU) ===")
    try:
        env = gym.make(ENV_NAME)
        env = ClipAction(env)
        env = gym.wrappers.RecordEpisodeStatistics(env)
        
        print("\nCreating optimized JAX model...")
        model = PPO_SBX(
            "MlpPolicy",
            env,
            verbose=0,
            **{k: v for k, v in SBX_CONFIG.items() if k != "policy_kwargs"},
            policy_kwargs=SBX_CONFIG["policy_kwargs"]
        )
        
        print("\nRunning JIT compilation...")
        model.learn(total_timesteps=1024)
        
        print("\nStarting main training...")
        start_time = time.time()
        model.learn(total_timesteps=TOTAL_TIMESTEPS)
        training_time = time.time() - start_time
        
        model.save(SBX_MODEL_PATH)
        print(f"\nSBX training completed in {training_time:.2f}s")
        return training_time
        
    except Exception as e:
        print(f"\n❌ SBX training failed: {type(e).__name__}: {e}")
        print("\nDebugging info:")
        if hasattr(e, 'args'):
            print("Exception args:", e.args)
        return None
    finally:
        env.close()

# ===== Results Handling =====
def save_results(sb3_time=None, sbx_time=None):
    results = {}
    if os.path.exists(RESULTS_FILE):
        with open(RESULTS_FILE, "r") as f:
            results = json.load(f)
    
    if sb3_time is not None:
        results["sb3_time"] = sb3_time
    if sbx_time is not None:
        results["sbx_time"] = sbx_time
    
    with open(RESULTS_FILE, "w") as f:
        json.dump(results, f, indent=2)

def load_results():
    if os.path.exists(RESULTS_FILE):
        with open(RESULTS_FILE, "r") as f:
            return json.load(f)
    return {"sb3_time": None, "sbx_time": None}

def run_comparison():
    """Run comparison with maximum optimizations"""
    results = load_results()
    
    # SB3 Training
    if not os.path.exists(SB3_MODEL_PATH) or force:
        results["sb3_time"] = train_sb3()
        save_results(sb3_time=results["sb3_time"])
    
    # SBX Training
    if not os.path.exists(SBX_MODEL_PATH) or force:
        results["sbx_time"] = train_sbx()
        save_results(sbx_time=results["sbx_time"])
    
    # Display results
    if results.get("sb3_time") and results.get("sbx_time"):
        print("\n=== Final Results ===")
        print(f"SB3 (PyTorch CPU): {results['sb3_time']:.2f}s")
        print(f"SBX (JAX CPU): {results['sbx_time']:.2f}s")
        
        speedup = results['sb3_time'] / results['sbx_time']
        print(f"Speedup: {speedup:.2f}x")
        
        # Enhanced visualization
        plt.figure(figsize=(10, 6))
        bars = plt.bar(["PyTorch CPU", "JAX CPU"], 
                      [results["sb3_time"], results["sbx_time"]],
                      color=['#3498db', '#e74c3c'])
        
        plt.ylabel("Training Time (seconds)", fontsize=12)
        plt.title(f"PPO Performance Comparison\n({TOTAL_TIMESTEPS:,} steps on {ENV_NAME})", 
                 fontsize=14, pad=20)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        
        # Add value labels
        for bar in bars:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height,
                    f'{height:.1f}s',
                    ha='center', va='bottom')
        
        # Add speedup annotation
        plt.annotate(f"{speedup:.2f}x faster",
                    xy=(1, results["sbx_time"]),
                    xytext=(10, 0), 
                    textcoords="offset points",
                    ha='left', va='center',
                    fontsize=12, color='green')
        
        plt.tight_layout()
        plt.savefig("ppo_comparison.png", dpi=300)
        plt.show()

if __name__ == "__main__":
    run_comparison()
```

```python
import os
import time
import json
import gymnasium as gym
import torch
import jax
import optax
import numpy as np
import matplotlib.pyplot as plt
from jax import numpy as jnp
from jax import vmap, pmap
from flax import linen as jnn
from stable_baselines3 import PPO as PPO_SB3
from sbx.ppo import PPO as PPO_SBX  # Ensure this import is correct
from jax import config
from gymnasium.wrappers import ClipAction

# ===== Fix numpy bool8 issue =====
np.bool8 = np.bool_  # Workaround for legacy numpy versions

# ===== JAX Configuration =====
config.update("jax_platform_name", "cpu")
config.update("jax_disable_jit", False)
config.update("jax_debug_nans", False)
print("Using JAX CPU backend with optimizations")

# ===== PyTorch Configuration =====
torch.set_num_threads(os.cpu_count())
print(f"PyTorch using {os.cpu_count()} threads")

# ===== Environment Configuration =====
ENV_NAME = "Pendulum-v1"
TOTAL_TIMESTEPS = 500_000

# File paths
RESULTS_FILE = "ppo_timing_results.json"

# ===== Fixed Hyperparameters =====
SBX_CONFIG = {
    "learning_rate": 3e-4,
    "gamma": 0.99,
    "gae_lambda": 0.95,
    "clip_range": 0.2,
    "batch_size": 2048,
    "n_steps": 4096,
    "n_epochs": 5,
    "policy_kwargs": {
        "net_arch": [128, 128],
        "activation_fn": jnn.relu,
        "ortho_init": False,
        "optimizer_class": optax.adamw,
        "optimizer_kwargs": {"eps": 1e-5, "weight_decay": 1e-4},
    }
}

def train_sb3():
    print("\n=== Training SB3 (PyTorch CPU) ===")
    env = gym.make(ENV_NAME)
    print("Action Space:", env.action_space)

    #env = ClipAction(env)
    env = gym.wrappers.RecordEpisodeStatistics(env)
    
    model = PPO_SB3("MlpPolicy", env, verbose=0, batch_size=2048)

    start_time = time.time()
    model.learn(total_timesteps=TOTAL_TIMESTEPS)
    sb3_total_time = time.time() - start_time

    print(f"SB3 Total Training Time: {sb3_total_time:.2f}s")
    env.close()
    return sb3_total_time

def train_sbx():
    print("\n=== Training SBX (JAX CPU) ===")
    env = gym.make(ENV_NAME)
    print("Action Space:", env.action_space)

    #env = ClipAction(env)
    env = gym.wrappers.RecordEpisodeStatistics(env)

    model = PPO_SBX(
        "MlpPolicy",
        env,
        verbose=0,
        **{k: v for k, v in SBX_CONFIG.items() if k != "policy_kwargs"},
        policy_kwargs=SBX_CONFIG["policy_kwargs"]
    )

    # Measure pre-warmup loop
    start_pre_warmup = time.time()
    model.learn(total_timesteps=TOTAL_TIMESTEPS // 10)
    pre_warmup_time = time.time() - start_pre_warmup

    # Measure post-warmup loop
    start_post_warmup = time.time()
    model.learn(total_timesteps=TOTAL_TIMESTEPS // 10)
    post_warmup_time = time.time() - start_post_warmup

    # Measure full SBX training
    start_total = time.time()
    model.learn(total_timesteps=TOTAL_TIMESTEPS)
    sbx_total_time = time.time() - start_total

    print(f"SBX Pre-Warmup Time: {pre_warmup_time:.2f}s")
    print(f"SBX Post-Warmup Time: {post_warmup_time:.2f}s")
    print(f"SBX Total Training Time: {sbx_total_time:.2f}s")

    env.close()
    return pre_warmup_time, post_warmup_time, sbx_total_time

def save_results(sb3_time=None, sbx_pre=None, sbx_post=None, sbx_total=None):
    results = {}
    if os.path.exists(RESULTS_FILE):
        with open(RESULTS_FILE, "r") as f:
            try:
                results = json.load(f)
            except json.JSONDecodeError:
                results = {}  # Initialize if the file is corrupted

    if sb3_time is not None:
        results["sb3_total_time"] = sb3_time
    if sbx_pre is not None:
        results["sbx_pre_warmup_time"] = sbx_pre
    if sbx_post is not None:
        results["sbx_post_warmup_time"] = sbx_post
    if sbx_total is not None:
        results["sbx_total_time"] = sbx_total

    with open(RESULTS_FILE, "w") as f:
        json.dump(results, f, indent=2)

def load_results():
    if os.path.exists(RESULTS_FILE):
        with open(RESULTS_FILE, "r") as f:
            try:
                return json.load(f)
            except json.JSONDecodeError:
                return {}  # Return empty if the file is corrupted
    return {}

def plot_results(results):
    labels = ["SB3 Total", "SBX Pre-Warmup", "SBX Post-Warmup", "SBX Total"]
    times = [
        results.get("sb3_total_time", 0),
        results.get("sbx_pre_warmup_time", 0),
        results.get("sbx_post_warmup_time", 0),
        results.get("sbx_total_time", 0)
    ]
    
    plt.figure(figsize=(8, 5))
    plt.bar(labels, times, color=['blue', 'orange', 'green', 'red'])
    plt.xlabel("Training Stages")
    plt.ylabel("Time (seconds)")
    plt.title("SB3 vs SBX Training Times")
    plt.show()

# Use vmap for vectorizing operations
def vectorized_operation(inputs):
    """Example of using vmap for batch processing."""
    def process_input(input_value):
        return input_value ** 2  # Example operation: squaring the input value
    return vmap(process_input)(inputs)

# Use pmap for parallel processing across multiple devices (e.g., multiple GPUs or TPUs)
def parallelized_operation(inputs):
    """Example of using pmap for parallel processing across devices."""
    def process_input(input_value):
        return input_value ** 2  # Example operation: squaring the input value
    return pmap(process_input)(inputs)

def run_comparison():
    """Run full comparison and timing measurements."""
    results = load_results()

    # SB3 Training
    if not results.get("sb3_total_time"):
        results["sb3_total_time"] = train_sb3()
        save_results(sb3_time=results["sb3_total_time"])

    # SBX Training
    if not results.get("sbx_total_time"):
        results["sbx_pre_warmup_time"], results["sbx_post_warmup_time"], results["sbx_total_time"] = train_sbx()
        save_results(sbx_pre=results["sbx_pre_warmup_time"],
                     sbx_post=results["sbx_post_warmup_time"],
                     sbx_total=results["sbx_total_time"])

    # Display results
    if results.get("sb3_total_time") and results.get("sbx_total_time"):
        print("\n=== Final Results ===")
        print(f"SB3 (PyTorch CPU) Total Time: {results['sb3_total_time']:.2f}s")
        print(f"SBX (JAX CPU) Pre-Warmup Time: {results['sbx_pre_warmup_time']:.2f}s")
        print(f"SBX (JAX CPU) Post-Warmup Time: {results['sbx_post_warmup_time']:.2f}s")
        print(f"SBX (JAX CPU) Total Time: {results['sbx_total_time']:.2f}s")
        
        plot_results(results)

if __name__ == "__main__":
    run_comparison()
```
