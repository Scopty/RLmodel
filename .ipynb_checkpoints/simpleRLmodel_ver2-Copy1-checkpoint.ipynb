```python
# TradingEnv
debug = False

class TradingEnv(Env):
    def __init__(self, df):
        super(TradingEnv, self).__init__()
        self.df = df
        self.current_step = 0
        self.initial_balance = 10000  # Starting cash
        self.balance = self.initial_balance
        self.net_worth = self.initial_balance
        self.max_net_worth = self.initial_balance
        self.shares = 0
        self.buy_price = 0
        self.total_reward = 0
        self.position_open = False  # Track open positions
        self.round_trip_trades = 0  # Counter for buy-sell cycles
        self.max_steps = 0
        reward = 0

        # Action Space: 0 = Hold, 1 = Buy, 2 = Sell
        self.action_space = Discrete(3)

        # ohlcv, balance, net worth, shares held, current step
        self.obs_shape = 12  # 9 original features + 3 mask elements
        self.observation_space = Box(
            low=-np.inf, high=np.inf, shape=(self.obs_shape,), dtype=np.float32
        )

    def get_action_mask(self) -> np.ndarray:
        """
        Returns a binary mask indicating valid actions.
        1 indicates a valid action, 0 indicates an invalid action.
        """
        action_mask = np.zeros(self.action_space.n, dtype=np.float32)
        for act in self.get_valid_actions():
            action_mask[act] = 1.0
        return action_mask

    def get_action_mask(self):
        # Example: Return a mask for 3 actions (0, 1, 2)
        valid_actions = [1, 0, 1]  # Allow actions 0 and 2
        return np.array(valid_actions, dtype=np.float32)  # Shape: (3,)
    
    def get_valid_actions(self):
        valid_actions = [0]  # Hold is always valid
        if not self.position_open:  # Allow Buy only if no position is open
            valid_actions.append(1)
        if self.position_open:  # Allow Sell only if a position is open
            valid_actions.append(2)
        return valid_actions

    def get_obs(self):

        # Include the current position
        obs = np.concatenate([
            self.df.iloc[self.current_step][["open", "high", "low", "close", "volume"]].values.astype(np.float64),  # (5,)
            np.array([self.shares], dtype=np.float64),   # (1,)
            np.array([self.balance], dtype=np.float64),  # (1,)
            np.array([self.net_worth], dtype=np.float64),# (1,)
            np.array([self.current_step], dtype=np.float64)  # (1,)
        ])
    
          # Get the action mask (for instance, shape: (3,))
        mask = self.get_action_mask()  # This should return an array like [1, 0, 1]
        
        # Concatenate the observation with the mask (resulting shape: (9+3=12,))
        combined_obs = np.concatenate([obs, mask])
        
        if debug: 
            print("Observation with mask:", combined_obs)

        # Debugging output
        if debug:
            print(f"Step {self.current_step}:")
            print(f"  - Original Observation (shape {obs.shape}): {obs}")
            print(f"  - Action Mask (shape {mask.shape}): {mask}")
            print(f"  - Combined Observation (shape {combined_obs.shape}): {combined_obs}")

        
        return combined_obs
    
    def reset(self, seed=None, options=None):
        super().reset(seed=seed)
        self.current_step = 0
        self.balance = self.initial_balance
        self.net_worth = self.initial_balance
        self.max_net_worth = self.initial_balance
        self.shares = 0
        self.position_open = False
        self.round_trip_trades = 0
        self.buy_price = 0
        self.max_steps = 0
        self.total_reward = 0
        reward = 0
        obs = self.get_obs()  # Use the updated observation method

        if debug:
            print(f"Environment reset: Initial Observation (shape {obs.shape}): {obs}")
        
        return obs, info

    def step(self, action): 
        if debug: print("Current Step:", self.current_step)
        
        if self.current_step >= len(self.df) - 1:
            return np.zeros(self.observation_space.shape), 0, True, False, {}
    
        close_price = self.df.iloc[self.current_step]["close"]
        if debug: print(' Close_price:', close_price)
        
        reward = 0
        done = False
        truncated = False
        valid_actions = self.get_valid_actions()
        
        # Create an action mask (1 for valid, 0 for invalid actions)
        action_mask = np.zeros(3, dtype=np.float32)
        for act in valid_actions:
            action_mask[act] = 1.0

        if debug: print(' Action', action)

        if action == 1:  # Buy
            self.shares = 1000
            self.balance -= close_price*self.shares
            self.position_open = True
            self.buy_price = close_price  # Store the buy price
            
        elif action == 2:  # Sell
            self.balance += close_price*self.shares
            reward = (close_price - self.buy_price)*self.shares  # Reward profit
            if debug: print('Trade profit (PnL):', reward)
            self.shares = 0
            self.buy_price = 0
            self.position_open = False
            self.round_trip_trades += 1

        # Update the total reward for the episode
        self.total_reward += reward ** 3
    
        # Calculate the net worth
        self.net_worth = self.balance + (self.shares * close_price)

        # Update the max net worth
        self.max_net_worth = max(self.net_worth, self.max_net_worth)

        if debug: print(" Market Data:", self.df.iloc[self.current_step][["open", "high", "low", "close", "volume"]].values)
        if debug: print(" Shares:", self.shares)
        if debug: print(" Balance:", self.balance)
        if debug: print(" Market position:", self.shares*close_price)
        if debug: print(" Net Worth:", self.net_worth)
            
        # Terminate after 10 round-trip trades or end of data
        if self.round_trip_trades >= 10:
            done = True
        if self.current_step >= len(self.df) - 1:
            done = True
        # Check if the episode is done
        if self.net_worth <= 0:
            done = True

        if debug: print(" Max Steps:", self.max_steps)
            
        self.current_step += 1
        
        if debug: print(f"Step: {self.current_step}, Valid Actions: {valid_actions}, Action Mask: {action_mask}, Selected Action: {original_action}, Invalid Action: {str(original_action not in valid_actions)}")
        obs = self.get_obs()  # Use the updated observation method
        info = {"valid_actions": valid_actions, "action_mask": action_mask}  # Include action mask in info

        if debug: 
            print(f"  - New Observation (shape {obs.shape}): {obs}")
            print(f"  - Action Mask (shape {action_mask.shape}): {action_mask}")
        
        return obs, reward, done, truncated, info
```

```python
import numpy as np

class ActionMasker(gym.Wrapper):
    """
    Wrapper for action masking in environments.
    Adds action mask as a part of the environment step.
    """
    def __init__(self, env: gym.Env, mask_fn: callable):
        super().__init__(env)
        self.mask_fn = mask_fn

    def step(self, action):
        obs, reward, done, truncated, info = self.env.step(action)
        
        # Get the action mask
        action_mask = self.mask_fn(self.env)
        
        # Add the action mask to the info dictionary
        info['action_mask'] = action_mask
        
        return obs, reward, done, truncated, info
    
    def reset(self, **kwargs):
        obs = self.env.reset(**kwargs)
        return obs
```

```python
from stable_baselines3.common.policies import ActorCriticPolicy
import torch
import torch.nn.functional as F

class MaskedPPOPolicy(ActorCriticPolicy):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)

    def forward(self, combined_obs, action_mask=None, deterministic=False):
        if debug: print(f"\n[MaskedPPOPolicy] Received Observation Shape: {combined_obs.shape}")
    
        # Ensure combined_obs has expected shape
        if combined_obs.shape[-1] != 12:
            raise ValueError(f"Expected observation shape (batch_size, 12), but got {combined_obs.shape}")
    
        # Use the entire observation as input (shape: [batch_size, 12])
        latent_pi, latent_vf = self.mlp_extractor(combined_obs)  # No need to split into obs/mask
    
        # Compute action distribution and value
        distribution = self._get_action_dist_from_latent(latent_pi)
        values = self.value_net(latent_vf)
    
        # Ensure action_mask is used properly
        if action_mask is None:
            action_mask = combined_obs[:, 9:]  # Extract last 3 elements as action mask
    
        if action_mask is not None:
            action_mask_tensor = torch.as_tensor(action_mask, dtype=torch.float32, device=combined_obs.device)
        
            # Ensure action mask has the same shape as logits
            action_mask_tensor = action_mask_tensor.view(-1, distribution.distribution.logits.shape[-1])
        
            # Apply masking to logits
            distribution.distribution.logits = distribution.distribution.logits.masked_fill(action_mask_tensor == 0, -1e9)

    
        # Action selection
        if deterministic:
            # Greedy action selection (choose the action with the highest probability)
            actions = torch.argmax(distribution.distribution.probs, dim=-1)
        else:
            # Sample action from the distribution
            actions = distribution.sample()
        log_probs = distribution.log_prob(actions)
    
        if debug: print(f"  - Selected Actions: {actions}")
    
        return actions, values, log_probs


    def predict(self, observation, state=None, mask=None, action_mask=None, deterministic=False):
        with torch.no_grad():
            observation = torch.as_tensor(observation, device=self.device, dtype=torch.float32)
            actions, _, _ = self.forward(observation, action_mask=action_mask, deterministic=deterministic)
            actions = actions.cpu().numpy()
        return actions, state
```

```python
# Define the mask_fn to get valid actions from the environment
def mask_fn(env: gym.Env) -> np.ndarray:
    return env.get_action_mask()  # Get valid action mask from the environment
```

```python
debug = False

from stable_baselines3.common.callbacks import BaseCallback
import numpy as np

class RewardCallback(BaseCallback):
    def __init__(self, verbose=0, debug=False):
        super(RewardCallback, self).__init__(verbose)
        self.debug = debug
        
        # Metrics for tracking actions and rewards
        self.episode_rewards = []
        self.episode_steps = []  # List to track steps per episode
        self.iteration_rewards = []
        self.iteration_invalid_actions = []
        self.invalid_actions = []  # Store invalid actions for debugging
        self.valid_actions = []
        self.current_episode_steps = 0  # Track steps in the current episode
        
        # Metrics for TensorBoard logging
        self.total_reward = 0
        self.reward = 0
        self.num_trades = 0

    def _on_step(self) -> bool:
        # Collect rewards and actions
        rewards = self.locals.get("rewards")
        actions = self.locals.get("actions")
        
        if rewards is not None:
            self.episode_rewards.extend(rewards)
        if actions is not None:
            # Check for invalid actions
            valid_actions = self.locals.get("infos")[0].get("valid_actions", [0, 1, 2])
            if actions[0] not in valid_actions:
                self.invalid_actions.append(actions[0])
            if actions[0] in valid_actions:
                self.valid_actions.append(actions[0])

        # Increment step counter
        self.current_episode_steps += 1

        # Access the environment properly
        if isinstance(self.training_env, DummyVecEnv):
            # For vectorized environments (like DummyVecEnv)
            env = self.training_env.envs[0]  # Get first environment instance
        else:
            env = self.training_env
        
        # Unwrap through ActionMasker to get the TradingEnv
        if hasattr(env, 'env'):  # ActionMasker wrapper
            env = env.env
        elif hasattr(env, 'envs'):  # If there's another VecEnv layer
            env = env.envs[0].env

        # Log custom metrics (total reward, reward, num_trades) for TensorBoard
        self.total_reward += env.total_reward  # Assuming total_reward is in the inner environment
        #self.reward += rewards
        self.num_trades += env.round_trip_trades  # Assuming `round_trip_trades` tracks trades

        # Log to TensorBoard
        self.logger.record("custom/num_trades", self.num_trades)
        self.logger.record("custom/total_reward", self.total_reward)
        #self.logger.record("custom/reward", self.reward)
        
        # Check if gradients are available for logging
        if self.locals.get("gradients_available", False):  # Check if gradients are available
            total_norm = 0
            for name, param in self.model.policy.named_parameters():
                if param.grad is not None:
                    param_norm = param.grad.norm(2).item()  # L2 norm of gradients
                    total_norm += param_norm ** 2
                    self.logger.record(f"gradients/{name}_norm", param_norm)
            total_norm = total_norm ** 0.5  # Total gradient norm for the model
            self.logger.record("gradients/total_norm", total_norm)

        # Log entropy of the policy
        if hasattr(self.model.policy, "action_dist"):
            action_dist = self.model.policy.action_dist
            entropy = action_dist.entropy().mean().item()  # Mean entropy over the batch
            self.logger.record("policy/entropy", entropy)
        elif hasattr(self.model.policy, "get_distribution"):
            action_dist = self.model.policy.get_distribution(self.locals.get("obs"))
            entropy = action_dist.entropy().mean().item()  # Mean entropy over the batch
            self.logger.record("policy/entropy", entropy)

        # Log Loss of Value Network
        if "value_loss" in self.locals:
            value_loss = self.locals["value_loss"]
            self.logger.record("loss/value_loss", value_loss)

        # Check if the episode is done
        dones = self.locals.get("dones")
        if dones is not None and any(dones):
            # Store the steps for the episode
            self.episode_steps.append(self.current_episode_steps)

            # Reset step counter for next episode
            self.current_episode_steps = 0

            # Calculate total reward for the episode
            total_reward = np.sum(self.episode_rewards)
            self.iteration_rewards.append(total_reward)
            self.episode_rewards = []

            # Count invalid and valid actions
            invalid_action_count = len(self.invalid_actions)
            valid_action_count = len(self.valid_actions)

            # Append the count of invalid actions for this iteration
            self.iteration_invalid_actions.append(invalid_action_count)

            # Print debug information if debug flag is set
            if self.debug:
                print(f"Invalid actions in this episode: {invalid_action_count}")
                print(f"Valid actions in this episode: {valid_action_count}")
                print(f"Invalid actions in this episode: {self.invalid_actions}")

            # Reset invalid actions for the next episode
            self.invalid_actions = []

        return True


# Initialize the callback
reward_callback = RewardCallback()
```

```python
# Test function

def test_agent(model, env, buy_signals_list, sell_signals_list, seed=42):
    """
    Test the trained model with action masking.
    
    Args:
        model: Trained PPO model.
        env: Environment with action masking.
        buy_signals_list: List to store buy signal timestamps.
        sell_signals_list: List to store sell signal timestamps.

    Returns:
        total_rewards: Total rewards accumulated during the test.
    """
    torch.manual_seed(seed)  # Ensure deterministic behavior
    
    # Access the environment wrapped by DummyVecEnv and ActionMasker
    env_inner = env.envs[0]  # Get the environment wrapped by DummyVecEnv (the first one)
    
    # Now we access the original TradingEnv wrapped inside ActionMasker
    env_inner = env_inner.env  # Access the original environment inside ActionMasker
    
    obs, _ = env_inner.reset()  # Reset the environment, now returns only the observation
    done = False
    total_rewards = 0

    while not done:
        for i in range(30):  # Run for a fixed number of steps

            # Extract the action mask from the original environment
            action_mask = env_inner.get_action_mask()  # Access get_action_mask from TradingEnv inside ActionMasker

            # Convert observation and mask to tensors
            obs_tensor = torch.as_tensor(obs, dtype=torch.float32, device=model.device).unsqueeze(0)
            action_mask_tensor = torch.as_tensor(action_mask, dtype=torch.float32, device=model.device).unsqueeze(0)

            # Predict action with action masking
            with torch.no_grad():
                action, _, _ = model.policy.forward(combined_obs=obs_tensor, action_mask=action_mask_tensor, deterministic=True)

            # Extract action probabilities correctly
            #action_probs = distribution.probs.detach().cpu().numpy()

            # Take the most probable action (ensures determinism)
            #action = np.argmax(action_probs)
            
            # Take the most probable action (ensures determinism)
            #action = torch.argmax(distribution.distribution.probs, dim=-1).cpu().numpy()[0]
            
            action = action.cpu().numpy()[0]  # Convert action tensor to numpy

            # Step the environment
            obs, reward, done, truncated, info = env_inner.step(action)

            # Debugging Output
            print(f"Step {i+1}: Action {action} | Reward: {reward:.2f} | Done: {done}")

            total_rewards += reward

            # Record buy or sell signals
            current_step = env_inner.current_step  # Access current_step directly from env_inner
            if action == 1:  # Buy signal
                buy_signals_list.append(df.index[current_step])
            elif action == 2:  # Sell signal
                sell_signals_list.append(df.index[current_step])

            if done or truncated:  # If the episode ends, reset the environment
                print(f" Total Reward: {total_rewards:.2f}")
                obs = env_inner.reset()  # Reset environment, which only returns the observation
                break  # Exit loop if episode ends

    return total_rewards
```

```python
# Produce grid results

def evaluate_model(model, env, num_episodes=10):
    total_rewards = []
    for _ in range(num_episodes):
        # Access the environment wrapped by DummyVecEnv and ActionMasker
        env_inner = env.envs[0]  # Get the environment wrapped by DummyVecEnv (the first one)
        # Now we access the original TradingEnv wrapped inside ActionMasker
        env_inner = env_inner.env  # Access the original environment inside ActionMasker
        
        obs, info = env_inner.reset()  # Reset the environment
        episode_reward = 0
        done = False

        while not done:
            for i in range(30):  # Run for a fixed number of steps
                # Extract the action mask from the original environment
                action_mask = env_inner.get_action_mask()  # Access get_action_mask from TradingEnv inside ActionMasker
    
                # Convert observation and mask to tensors
                obs_tensor = torch.as_tensor(obs, dtype=torch.float32, device=model.device).unsqueeze(0)
                action_mask_tensor = torch.as_tensor(action_mask, dtype=torch.float32, device=model.device).unsqueeze(0)
    
                # Predict action with action masking
                with torch.no_grad():
                    action, _, _ = model.policy.forward(obs_tensor, action_mask=action_mask_tensor, deterministic=True)
    
                action = action.cpu().numpy()[0]  # Convert action tensor to numpy
    
                # Step the environment
                obs, reward, done, truncated, info = env_inner.step(action)
    
                episode_reward += reward

                if done or truncated:  # If the episode ends, reset the environment
                    print(f" Total Reward: {episode_reward:.2f}")
                    obs = env_inner.reset()  # Reset environment, which only returns the observation
                    break  # Exit loop if episode ends

        total_rewards.append(episode_reward)

    return np.mean(total_rewards)

from itertools import product

# Define hyperparameter ranges
ent_coef_range = [0.01]       # Lower values for stability
gamma_range = [0.9]              # Lower values for short-term focus
gae_lambda_range = [0.95]        # Wider range to test variance
n_steps_range = [2048]           # Larger values for more experiences
clip_range_range = [0.1]         # Narrower range for stability
n_epochs_range = [3]               # Fewer epochs for faster training
batch_size_range = [64]           # Larger batches for better performance

# Perform grid search
results = []
models = []

for ent_coef, gamma, gae_lambda, n_steps, clip_range, n_epochs, batch_size in product(
    ent_coef_range, gamma_range, gae_lambda_range, n_steps_range, clip_range_range, n_epochs_range, batch_size_range
):
    print(f"Training model with: ent_coef={ent_coef}, gamma={gamma}, gae_lambda={gae_lambda}, "
          f"n_steps={n_steps}, clip_range={clip_range}, n_epochs={n_epochs}, batch_size={batch_size}")
    
    model = PPO(
        MaskedPPOPolicy,
        env,
        ent_coef=ent_coef,
        gamma=gamma,
        gae_lambda=gae_lambda,
        n_steps=n_steps,
        clip_range=clip_range,
        n_epochs=n_epochs,
        batch_size=batch_size,
        verbose=1,
        tensorboard_log="./tensorboard_logs/"
    )
    
    # Train for 100,000 timesteps
    model.learn(total_timesteps=100000)
    
    # Store results (e.g., total reward)
    total_reward = evaluate_model(model, env)  # Define a function to evaluate the model

    # Store the model and its parameters
    models.append({
        "model": model,  # Store the trained model
        "params": {
            "ent_coef": ent_coef,
            "gamma": gamma,
            "gae_lambda": gae_lambda,
            "n_steps": n_steps,
            "clip_range": clip_range,
            "n_epochs": n_epochs,
            "batch_size": batch_size,
        }
    })
    
    results.append({
        "ent_coef": ent_coef,
        "gamma": gamma,
        "gae_lambda": gae_lambda,
        "n_steps": n_steps,
        "clip_range": clip_range,
        "n_epochs": n_epochs,
        "batch_size": batch_size,
        "total_reward": total_reward
    })

# Save results to a CSV file
import pandas as pd
results_df = pd.DataFrame(results)
name = "grid_search_results_optimize_row_9_100000_steps.csv"
results_df.to_csv(name, index=False)

print("Grid search completed. Results saved to", name)
```

```python
import mplfinance as mpf
import matplotlib.dates as mdates
import pandas as pd
import matplotlib.pyplot as plt

# Function to calculate profit based on buy and sell signals
def calculate_profit(buy_signals, sell_signals, df):
    buy_prices = df.loc[buy_signals, 'close']
    sell_prices = df.loc[sell_signals, 'close']

    # Ensure that the number of buy signals is less than or equal to the number of sell signals
    min_length = min(len(buy_prices), len(sell_prices))

    # Calculate profit for each pair of buy and sell signals
    profits = []
    for i in range(min_length):
        profit = sell_prices.iloc[i] - buy_prices.iloc[i]  # Profit = Sell Price - Buy Price
        profits.append(profit)

    total_profit = sum(profits)
    return total_profit, profits

# Function to plot the signals and compare profits
def plot_signals_and_compare_profit(buy_signals, sell_signals, df, title):
    # Convert buy/sell signals to DatetimeIndex
    buy_signals = pd.to_datetime(buy_signals)  
    buy_signals = buy_signals[buy_signals.isin(df.index)]  # Filter buy signals to match df.index

    sell_signals = pd.to_datetime(sell_signals)  
    sell_signals = sell_signals[sell_signals.isin(df.index)]  # Filter sell signals to match df.index

    # Create new columns to mark the buy and sell signals with NaN for non-signals
    buy_signal_prices = df['close'].copy()
    buy_signal_prices[~df.index.isin(buy_signals)] = float('nan')  # Set non-buy signals to NaN

    sell_signal_prices = df['close'].copy()
    sell_signal_prices[~df.index.isin(sell_signals)] = float('nan')  # Set non-sell signals to NaN

    # Create addplots for buy and sell signals
    buy_signal_plot = mpf.make_addplot(buy_signal_prices, type='scatter', markersize=40, marker='^', color='green')
    sell_signal_plot = mpf.make_addplot(sell_signal_prices, type='scatter', markersize=40, marker='v', color='red')

    # Plot with the added buy and sell signals
    fig, axes = mpf.plot(
        df,
        type='ohlc',
        style='charles',
        datetime_format='%Y-%m-%d %H:%M',
        addplot=[buy_signal_plot, sell_signal_plot],
        returnfig=True,
        figsize=(20, 10),
        warn_too_much_data=1000  # Adjust the size again to maintain consistency
    )

    # Customize x-axis
    #axes[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    #plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

    # Add legend manually
    #plt.legend(['Buy Signals', 'Sell Signals'], loc='upper left')

    # Add title
    #plt.title(title)

    # Show the plot
    #plt.show()

    # Calculate and print profit for the signals
    total_profit, profits = calculate_profit(buy_signals, sell_signals, df)
    print(f"Total Profit for {title}: {total_profit:.2f}")
    return total_profit

# Calculate and plot for 1000 timesteps signals
profit_1000 = plot_signals_and_compare_profit(buy_signals_1000, sell_signals_1000, df, title="Profit for 1000 Timesteps")

# Calculate and plot for 10000 timesteps signals
profit_10000 = plot_signals_and_compare_profit(buy_signals_10000, sell_signals_10000, df, title="Profit for 10000 Timesteps")

# Calculate and plot for 100000 timesteps signals
#profit_100000 = plot_signals_and_compare_profit(buy_signals_100000, sell_signals_100000, df, title="Profit for 100000 Timesteps")


# Compare the total profits
print(f"Total Profit for 1000 Timesteps: {profit_1000:.2f}")
print(f"Total Profit for 10000 Timesteps: {profit_10000:.2f}")
#print(f"Total Profit for 100000 Timesteps: {profit_100000:.2f}")
```
