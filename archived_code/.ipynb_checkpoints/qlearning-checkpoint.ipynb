```python
import yfinance as yf
import numpy as np
import gym
from gym import spaces
import matplotlib.pyplot as plt

class StockTradingEnv(gym.Env):
    def __init__(self, stock_data, initial_balance=10000):
        super(StockTradingEnv, self).__init__()
        
        self.action_space = spaces.Discrete(3)  # Buy, Sell, Hold
        self.observation_space = spaces.Box(low=0, high=1, shape=(6,), dtype=np.float32)
        
        self.stock_data = stock_data
        self.initial_balance = initial_balance
        self.balance = initial_balance
        self.shares_held = 0
        self.current_step = 0
        
        # Normalize only numeric columns (exclude 'Date' if present)
        numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
        self.stock_data_normalized = (
            self.stock_data[numeric_cols] - self.stock_data[numeric_cols].min()
        ) / (
            self.stock_data[numeric_cols].max() - self.stock_data[numeric_cols].min()
        )
        
    def reset(self):
        self.balance = self.initial_balance
        self.shares_held = 0
        self.current_step = 0
        return self._next_observation()
    
    def _next_observation(self):
        # Extract normalized features as a 1D numpy array
        current_row = self.stock_data_normalized.iloc[self.current_step].to_numpy(dtype=np.float32)
        
        # Normalize balance and reshape to match dimensions
        balance_norm = np.array([self.balance / self.initial_balance], dtype=np.float32).flatten()
        
        # Concatenate both arrays (ensuring both are 1D)
        obs = np.concatenate((current_row, balance_norm))
        
        return obs
    
    def step(self, action):
        # Ensure current_price is a float
        current_price = float(self.stock_data['Close'].iloc[self.current_step])
        
        if action == 0:  # Buy
            shares_to_buy = int(self.balance // current_price)
            self.shares_held += shares_to_buy
            self.balance -= shares_to_buy * current_price
        elif action == 1 and self.shares_held > 0:  # Sell
            self.balance += self.shares_held * current_price
            self.shares_held = 0
    
        # Move to next step
        self.current_step += 1
        done = self.current_step >= len(self.stock_data) - 1
    
        next_price = float(self.stock_data['Close'].iloc[self.current_step]) if not done else current_price
        portfolio_value = self.balance + self.shares_held * next_price
        reward = portfolio_value - self.initial_balance
    
        return self._next_observation(), reward, done, {}



# Download stock data
stock_data = yf.download('AAPL', start='2020-01-01', end='2023-01-01')
stock_data.reset_index(inplace=True)

env = StockTradingEnv(stock_data)

# Q-Learning parameters
num_episodes = 1000
alpha = 0.1
gamma = 0.99
epsilon = 1.0
epsilon_decay = 0.995
min_epsilon = 0.01

# Initialize Q-table
num_states = 10  # Simplified state space
num_actions = env.action_space.n
Q = np.zeros((num_states, num_actions))

# Discretize the state space
def discretize_state(state):
    return int(np.mean(state) * num_states) % num_states

# Q-Learning algorithm
for episode in range(num_episodes):
    state = env.reset()
    state = discretize_state(state)
    done = False
    
    while not done:
        if np.random.rand() < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[state])
        
        next_state, reward, done, _ = env.step(action)
        next_state = discretize_state(next_state)
        
        Q[state, action] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])
        
        state = next_state
    
    epsilon = max(min_epsilon, epsilon * epsilon_decay)
    
    if episode % 100 == 0:
        print(f'Episode {episode}, Epsilon: {epsilon}, Total Reward: {reward}')
```

```python
# Discretize the state space
def discretize_state(state):
    return int(np.mean(state) * num_states) % num_states


# Q-Learning algorithm
for episode in range(num_episodes):
    state = env.reset()
    state = discretize_state(state)
    done = False

    while not done:
        if np.random.rand() < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[state])

        next_state, reward, done, _ = env.step(action)
        next_state = discretize_state(next_state)

        Q[state, action] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])

        state = next_state

    epsilon = max(min_epsilon, epsilon * epsilon_decay)

    if episode % 100 == 0:
        print(f'Episode {episode}, Epsilon: {epsilon}, Total Reward: {reward}')

state = env.reset()
state = discretize_state(state)
done = False
total_reward = 0

while not done:
    action = np.argmax(Q[state])
    next_state, reward, done, _ = env.step(action)
    next_state = discretize_state(next_state)
    total_reward += reward
    state = next_state

print(f'Total Reward: {total_reward}')

plt.plot(stock_data['Close'])
plt.title('Stock Price')
plt.show()
```

```python
# Evaluate the learned policy
test_episodes = 10
for i in range(test_episodes):
    state = env.reset()
    state = discretize_state(state)
    done = False
    total_reward = 0  # In this setup, reward is the change in portfolio value.
    
    while not done:
        # Choose the best action (greedy) from the Q-table
        action = np.argmax(Q[state])
        next_state, reward, done, _ = env.step(action)
        state = discretize_state(next_state)
        total_reward = reward  # reward is calculated as portfolio_value - initial_balance
    
    print(f"Test Episode {i+1}: Total Reward (Profit) = {total_reward}")
```

```python
import matplotlib.pyplot as plt

episode_rewards = []

# Q-Learning algorithm (modified to track rewards)
for episode in range(num_episodes):
    state = env.reset()
    state = discretize_state(state)
    done = False
    total_episode_reward = 0
    
    while not done:
        if np.random.rand() < epsilon:
            action = env.action_space.sample()
        else:
            action = np.argmax(Q[state])
        
        next_state, reward, done, _ = env.step(action)
        next_state = discretize_state(next_state)
        
        Q[state, action] += alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])
        
        state = next_state
        total_episode_reward = reward  # In this case, reward is portfolio_value - initial_balance
    
    episode_rewards.append(total_episode_reward)
    epsilon = max(min_epsilon, epsilon * epsilon_decay)
    
    if episode % 100 == 0:
        print(f'Episode {episode}, Epsilon: {epsilon}, Total Reward: {total_episode_reward}')

# Plot the rewards over episodes
plt.plot(episode_rewards)
plt.xlabel("Episode")
plt.ylabel("Total Reward (Profit)")
plt.title("Q-Learning Training Performance")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Optionally: Save your Q-table for later use
np.save("q_table.npy", Q)

# Run a test episode with the trained Q-learning policy and record the portfolio value over time
state = env.reset()
state = discretize_state(state)
done = False

# Initialize a list to store portfolio values over the episode
portfolio_values = []
# Record the initial portfolio value
current_price = float(env.stock_data['Close'].iloc[env.current_step])
portfolio_value = env.balance + env.shares_held * current_price
portfolio_values.append(portfolio_value)

# Also, record the corresponding stock prices for reference
stock_prices = [current_price]

# Run the simulation until the episode is done
while not done:
    # Select the best action from the Q-table (no exploration)
    action = np.argmax(Q[state])
    
    next_state, reward, done, _ = env.step(action)
    state = discretize_state(next_state)
    
    # Get the new portfolio value
    current_price = float(env.stock_data['Close'].iloc[env.current_step])
    portfolio_value = env.balance + env.shares_held * current_price
    portfolio_values.append(portfolio_value)
    stock_prices.append(current_price)

# Print the final profit or loss
print(f'Total Reward (Profit/Loss): {portfolio_values[-1] - env.initial_balance}')

# Plot portfolio value vs. stock price
plt.figure(figsize=(12, 6))
plt.plot(portfolio_values, label='Portfolio Value')
plt.plot(stock_prices, label='Stock Price', alpha=0.6)
plt.xlabel("Time Step")
plt.ylabel("Value")
plt.title("Portfolio Value vs Stock Price Over Time")
plt.legend()
plt.show()
```
